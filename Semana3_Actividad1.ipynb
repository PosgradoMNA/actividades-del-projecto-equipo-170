##**Parte 1: Fundamentos de Bases de Datos**

###**Fundamentos de Bases de Datos para Ciencia de Datos**

Una base de datos se puede definir como un programa que almacena conjuntos de datos estructurados y relacionados, por ejemplo, la lista de clientes de alguna tienda o la lista de estudiantes de una escuela. 
Las bases de datos se componen de un elemento base llamado celda, el cual, en conjunto de otras celdas, forman las llamadas filas y columnas. Cuando las filas (también llamados registros) y las columnas (campos) se combinan, se forman las tablas. Por regla general en las tablas, la primera fila se reserva para los nombres del campo.



Ya que sabemos como se compone una base de datos, podemos describir cada uno de estos elementos antes mencionados:


Las **columnas (Campos)**, son las encargadas de dividir los datos en categorías, por ejemplo: Nombre, Apellido, Edad, etc.


Las **filas (Registros)**, recopilan la información relacionada dividida en columnas o campos.


Las **celdas**, son aquellos elementos que contienen la información de cada uno de los registros.


Los **datos**, hacen referencia a toda la información contenida en la base de datos.


###**Fundamentos de Almacenes de Datos (Data Warehouse) para Ciencia de Datos**

Un almacén de datos se define como un tipo de sistema manejador de datos diseñado para el análisis de negocios inteligentes. Básicamente es una base de datos enorme que contiene información de diferentes fuentes. Estos almacenes de datos resguardan múltiples bases de datos. Cada base de datos se organiza por tablas y columnas. Y dentro de sus ventajas se encuentra el mejoramiento en toma de decisiones, el análisis de datos históricos y, exactitud, calidad y consistencia en los datos.

Un almacén de datos debe contener los siguientes elementos:
* Bases de datos relacionales.
* Una solución ELT (Extraction, Loading and Transformation) para preparar los datos.
* Capacidades para el análisis estadístico, reportes y minado de datos.
* Herramientas para la visualización de datos para los usuarios.
* Capacidad para aplicar algoritmos de IA.
